```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dailyactivity = pd.read_csv('dailyActivity_merged.csv')
print('shape:',dailyactivity.shape)
dailyactivity.head()
```

```python
dailycalories=pd.read_csv('dailyCalories_merged.csv')
print('shape:',dailycalories.shape)
dailycalories.head()
```

```python
dailyintensities=pd.read_csv('dailyIntensities_merged.csv')
print('shape:',dailyintensities.shape)
dailyintensities.head()
```

```python
dailysteps=pd.read_csv('dailySteps_merged.csv')
print('shape:',dailysteps.shape)
dailysteps.head()
```

```python
heartrateseconds=pd.read_csv('heartrate_seconds_merged.csv')
print('shape:',heartrateseconds.shape)
heartrateseconds.head()
```

```python
sleepday=pd.read_csv('sleepDay_merged.csv')
print('shape:',sleepday.shape)
sleepday.head()
```

```python
weightinfo=pd.read_csv('weightLogInfo_merged.csv')
print('shape:',weightinfo.shape)
weightinfo.head()
```

```python
dailyactivity['ActivityDate'] = pd.to_datetime(dailyactivity['ActivityDate'])
dailycalories['ActivityDay'] = pd.to_datetime(dailycalories['ActivityDay'])
dailyintensities['ActivityDay'] = pd.to_datetime(dailyintensities['ActivityDay'])
dailysteps['ActivityDay'] = pd.to_datetime(dailysteps['ActivityDay'])
sleepday['SleepDay'] = pd.to_datetime(sleepday['SleepDay'])
weightinfo['Date'] = pd.to_datetime(weightinfo['Date'])
heartrateseconds['Time'] = pd.to_datetime(heartrateseconds['Time'])
```

```python
heartrate_daily = (
    heartrateseconds
    .groupby([heartrateseconds['Id'],
              heartrateseconds['Time'].dt.date])['Value']
    .mean()
    .reset_index()
)

heartrate_daily.columns = ['Id','ActivityDate','AvgHeartRate']
heartrate_daily['ActivityDate'] = pd.to_datetime(heartrate_daily['ActivityDate'])
```

```python
merged_df = dailyactivity.merge(
    dailycalories,
    left_on=['Id','ActivityDate'],
    right_on=['Id','ActivityDay'],
    how='left'
)
```

```python
merged_df = merged_df.merge(
    dailyintensities,
    on=['Id','ActivityDay'],
    how='left'
)
```

```python
merged_df = merged_df.merge(
    dailysteps,
    on=['Id','ActivityDay'],
    how='left'
)
```

```python
merged_df = merged_df.merge(
    sleepday,
    left_on=['Id','ActivityDate'],
    right_on=['Id','SleepDay'],
    how='left'
)




merged_df = merged_df.merge(
    weightinfo,
    left_on=['Id','ActivityDate'],
    right_on=['Id','Date'],
    how='left'
)
```

```python
merged_df = merged_df.merge(
    heartrate_daily,
    on=['Id','ActivityDate'],
    how='left'
)
merged_df = merged_df.drop(
    columns=['ActivityDate','SleepDay','Date'],
    errors='ignore'
)
merged_df.head()
merged_df.info()
```

```python
#L1. Classify Activity Level using Lambda

merged_df["Activity_Class"] = merged_df["TotalSteps"].apply(
    lambda x: "Low" if x < 5000 else ("Medium" if x < 10000 else "High")
)

merged_df[["TotalSteps", "Activity_Class"]].head()
```

```python
#L2. Classify Sleep Quality
merged_df["Sleep_Class"] = merged_df["TotalMinutesAsleep"].apply(
    lambda x: "Poor" if x < 360 else ("Good" if x <= 480 else "Excellent")
)
merged_df
```

```python
# L3. Calories Level Classification
merged_df["Calorie_Level"] = merged_df["Calories_x"].apply(
    lambda x: "High" if x > 2500 else "Normal"
)
merged_df
```

```python
#L4. Weekday vs Weekend Column

merged_df["DayType"] = merged_df["ActivityDay"].dt.dayofweek.apply(
    lambda x: "Weekend" if x >= 5 else "Weekday"
)
merged_df
```

```python
#L5. Steps per Active Minute

merged_df["Steps_per_ActiveMin"] = merged_df["TotalSteps"] / (
    merged_df["VeryActiveMinutes_x"] + merged_df["FairlyActiveMinutes_x"] + 1
)
merged_df
```

```python
# L6 Flag Sedentary Risk Users
merged_df["Sedentary_Risk"] = merged_df["TotalSteps"].apply(
    lambda x: "Yes" if x < 3000 else "No"
)
merged_df
```

```python
#l7 Convert Minutes to Hours
merged_df["Sleep_Hours"] = merged_df["TotalMinutesAsleep"].apply(lambda x: x/60)
merged_df
```

```python
#L8 Active vs Inactive Day Label
merged_df["Day_Status"] = merged_df["TotalSteps"].apply(
    lambda x: "Inactive" if x == 0 else "Active"
)
merged_df
```

```python
#F1. Function to Classify Activity Level
def activity_level(steps):
    if steps < 5000:
        return "Sedentary"
    elif steps < 10000:
        return "Moderate"
    else:
        return "Active"

merged_df["Activity_Level"] = merged_df["TotalSteps"].apply(activity_level)
merged_df
```

```python
#F2. BMI Calculation Function
def calculate_bmi(weight, height=1.75):
    return weight / (height ** 2)

merged_df["BMI"] = merged_df["WeightKg"].apply(
    lambda x: calculate_bmi(x) if pd.notnull(x) else None
)
calculate_bmi
merged_df
```

```python
#F3. Fitness Score Function

def fitness_score(row):
    return row["TotalSteps"] + (row["Calories_x"] / 10) + row["VeryActiveMinutes_x"]

merged_df["FitnessScore"] = merged_df.apply(fitness_score, axis=1)
merged_df
```

```python
#F4. Detect Inactive Days Function

def inactive_day(steps):
    return True if steps == 0 else False

merged_df["Inactive_Day"] = merged_df["TotalSteps"].apply(inactive_day)

merged_df
```
